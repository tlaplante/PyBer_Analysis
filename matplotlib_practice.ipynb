```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis,y_axis,marker="*",color="blue",linewidth=2,label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the plot

plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title

plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color="magenta", label='Boston')

# and axis labels
plt.xlabel("Fare($)")
plt.ylabel("Date")

# invert yaxis so Jan is on top
plt.gca().invert_yaxis()

# add a legend
plt.legend()

# add a title
plt.title("PyBer Fare by Month")

```

```python
# Create the plot with ax.plt() to create vertical bar chart

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() to create horizontal bar chart

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicage')

# and axis labels
plt.xlabel("Fare($)")
plt.ylabel("Date")

# invert yaxis so Jan is on top
plt.gca().invert_yaxis()

# add a legend
plt.legend()

# add a title
plt.title("PyBer Fare by Month")
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(y_axis,
           x_axis,
           s=[i*5 for i in y_axis],
           color = 'skyblue',
           alpha=.8,
           edgecolor='black',
           linewidths=2,
           label='Boston')

plt.legend()
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.xlim(0,50)
```

```python
# Create a Pie Chart Using the MATLAB Method
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
        colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"],
        labels=x_axis,
        explode=explode_values,
        autopct='%.1f%%')
```

```python
# Create a Pie Chart Using the Object-Oriented Interface

explode_values=(0, 0, 0.5, 0, 0, 0, 0.5, 0, 0, 0, 0, 0)
fig, ax = plt.subplots()

#for i in y_axis:
#    y_axis_larger=i*5

ax.pie(y_axis,
       labels=x_axis,
       colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"],
       autopct='%.1f%%',
#       figsize=(8,8),
       shadow=True,
       startangle=93,
       counterclock=False,
       radius=2,
       explode=explode_values
      )
plt.show()
```
